## 🔹 FacetGrid: Çoklu Grafik Panelleri

```python
g = sns.FacetGrid(df, col="sex", row="smoker")g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")g.set_titles(row_template="{row_name}", col_template="{col_name}")plt.suptitle("Cinsiyet ve Sigara Durumuna Göre Dağılım", y=1.03)plt.show()
```

## 🔹 pairplot: Sayısal Değişkenler Arası Dağılım

```python
sns.pairplot(df, hue="sex", diag_kind="kde")plt.suptitle("Sayısal Değişkenler Arası İlişki", y=1.02)plt.show()
```

## 🔹 catplot: Kategorik Veriler İçin Çok Amaçlı Grafik

```python
sns.catplot(x="day", y="total_bill", hue="sex", col="time",             kind="box", data=df, height=4, aspect=0.8)plt.suptitle("Zaman ve Gün Bazlı Fatura Dağılımı", y=1.05)plt.show()
```
