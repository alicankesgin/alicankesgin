## 🔹 Bar Chart (Çubuk Grafiği)

```python
kategoriler = ["A", "B", "C", "D"]degerler = [23, 45, 12, 36]plt.bar(kategoriler, degerler)plt.title("Kategoriye Göre Dağılım")plt.xlabel("Kategori")plt.ylabel("Değer")plt.show()
```

## 🔹 Histogram

```python
veri = np.random.randn(1000)plt.hist(veri, bins=30, edgecolor='black')plt.title("Normal Dağılım Histogramı")plt.xlabel("Değer")plt.ylabel("Frekans")plt.show()
```

## 🔹 Scatter Plot (Dağılım Grafiği)

```python
x = np.random.rand(100)y = np.random.rand(100)plt.scatter(x, y, alpha=0.7)plt.title("Rastgele Nokta Dağılımı")plt.xlabel("X")plt.ylabel("Y")plt.grid(True)plt.show()
```

## 🔹 Çoklu Grafik (subplot)

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))axs[0].plot(np.sin(x))axs[0].set_title("Sinüs")axs[1].plot(np.cos(x), color='orange')axs[1].set_title("Kosinüs")plt.suptitle("Sinüs ve Kosinüs Fonksiyonları")plt.tight_layout()plt.show()
```
