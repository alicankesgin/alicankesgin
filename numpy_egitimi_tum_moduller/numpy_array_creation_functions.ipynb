# NumPy Fonksiyonları: Array Oluşturma FonksiyonlarıBu bölümde NumPy ile dizi oluşturmak için kullanılan tüm fonksiyonlar detaylı açıklamalar ve örneklerle incelenecektir.

## 1. np.array()Python listeleri veya tuple'lardan NumPy dizisi (ndarray) oluşturmak için kullanılır.

```python
import numpy as np# Listeyi diziye çevirmea = np.array([1, 2, 3])print("1D dizi:", a)# Çok boyutlu yapıb = np.array([[1, 2], [3, 4]])print("2D dizi:\n", b)# Tuple'dan dizic = np.array((5, 6, 7))print("Tuple'dan dizi:", c)# dtype belirlemed = np.array([1, 2, 3], dtype=float)print("float dtype:", d)
```

## 2. np.zeros()Belirtilen şekle göre tüm elemanları sıfır olan bir NumPy dizisi oluşturur.

```python
z = np.zeros((2, 3))print("2x3 sıfırlardan oluşan dizi:\n", z)
```

## 3. np.ones()Tüm elemanları 1 olan dizi oluşturur.

```python
o = np.ones((3, 2))print("3x2 birlerden oluşan dizi:\n", o)
```

## 4. np.full()Tüm elemanları belirli bir sabit olan dizi oluşturur.

```python
f = np.full((2, 2), 9)print("2x2 dokuzlardan oluşan dizi:\n", f)
```

## 5. np.empty()İçeriği belirsiz, rastgele olan bir dizi oluşturur. Performans için sıfırlama yapmaz.

```python
e = np.empty((2, 2))print("Boş (rastgele içerikli) dizi:\n", e)
```

## 6. np.eye()Birim matris oluşturur. Diagonal 1, diğer elemanlar 0'dır.

```python
identity = np.eye(3)print("3x3 birim matris:\n", identity)
```

## 7. np.identity()np.eye() gibi ama sadece kare matrisler içindir.

```python
iden = np.identity(4)print("4x4 birim matris:\n", iden)
```

## 8. np.arange()Belirli bir başlangıç, bitiş ve adım değerine göre dizi oluşturur.

```python
ar = np.arange(0, 10, 2)print("0'dan 10'a kadar 2 artan dizi:", ar)
```

## 9. np.linspace()Başlangıç ve bitiş dahil olmak üzere eşit aralıklı sayılar üretir.

```python
lin = np.linspace(0, 1, 5)print("0 ile 1 arasında 5 eşit sayı:", lin)
```

## 10. np.logspace()Logaritmik olarak aralıklı sayılar üretir. Genellikle bilimsel hesaplamalarda kullanılır.

```python
log = np.logspace(1, 3, num=4)print("10^1 ile 10^3 arasında logaritmik aralıklı 4 sayı:", log)
```

## 12. np.fromiter()İterable (örneğin generator) üzerinden dizi oluşturur.

```python
it = (x*x for x in range(5))arr = np.fromiter(it, dtype=int)print("Generator'dan dizi:", arr)
```

## 13. np.asarray()Python list veya tuple’ı ndarray’e dönüştürür ama zaten ndarray ise kopya oluşturmaz.

```python
lst = [1, 2, 3]arr1 = np.asarray(lst)print("Liste'den asarray:", arr1)arr2 = np.asarray(arr1)print("Zaten ndarray olan:", arr2 is arr1)  # True, kopya oluşturmaz
```
