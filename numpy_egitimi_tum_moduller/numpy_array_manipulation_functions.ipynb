# NumPy Fonksiyonları: Dizi Şekillendirme (Array Manipulation)Bu bölümde NumPy dizilerinin yapısını değiştirmeye yönelik tüm fonksiyonlar örnekleriyle anlatılır.

## 1. reshape()Dizinin şeklini değiştirmek için kullanılır. Eleman sayısı aynı kalmalıdır.

```python
import numpy as npa = np.arange(12)b = a.reshape((3, 4))print("Orijinal dizi:", a)print("Yeniden şekillendirilmiş (3x4):\n", b)
```

## 2. ravel()Diziyi tek boyutlu hale getirir. Orijinal veriyi döndürür (view).

```python
b = np.array([[1, 2], [3, 4]])flat = b.ravel()print("Ravel ile tek boyutlu:", flat)
```

## 3. flatten()Diziyi tek boyutlu hale getirir ama her zaman bir kopya üretir.

```python
flat_copy = b.flatten()print("Flatten ile kopyalanmış tek boyutlu:", flat_copy)
```

## 4. resize()reshape gibidir ancak orijinal dizinin boyutunu kalıcı olarak değiştirir.

```python
a = np.array([[1, 2], [3, 4]])a.resize((1, 4))print("Resize sonrası dizi:", a)
```

## 5. transpose() ve TMatrisin satır ve sütunlarını yer değiştirir.

```python
a = np.array([[1, 2, 3], [4, 5, 6]])print("Orijinal:", a)print("Transpose:", a.transpose())print("Kısa yol (T):", a.T)
```

## 6. swapaxes()Belirtilen iki ekseni yer değiştirir (çok boyutlu dizilerde kullanılır).

```python
a = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])print("Orijinal shape:", a.shape)swapped = np.swapaxes(a, 0, 2)print("Yeni shape:", swapped.shape)
```

### reshape() vs resize() Farkı- `reshape()` geçici olarak yeni şekil verir, orijinali değiştirmez.- `resize()` kalıcı olarak değiştirir ve gerekirse veri ekler/kısaltır.

## 7. expand_dims()Belirli bir eksene yeni bir boyut ekler.

```python
a = np.array([1, 2, 3])b = np.expand_dims(a, axis=0)c = np.expand_dims(a, axis=1)print("Orijinal:", a.shape)print("Axis=0:", b.shape)print("Axis=1:", c.shape)
```

## 8. squeeze()Boyutu 1 olan eksenleri kaldırır.

```python
a = np.array([[[1], [2], [3]]])print("Orijinal shape:", a.shape)squeezed = np.squeeze(a)print("Sıkıştırılmış shape:", squeezed.shape)
```

## 9. rollaxis()Belirtilen ekseni istenen konuma taşır (az kullanılır).

```python
a = np.ones((3,4,5))print("Orijinal shape:", a.shape)rolled = np.rollaxis(a, 2, 0)print("Yeni shape:", rolled.shape)
```

## 10. moveaxis()Yeni NumPy sürümlerinde eksen taşımak için önerilen yöntemdir.

```python
a = np.zeros((3, 4, 5))b = np.moveaxis(a, 0, -1)print("Yeni shape:", b.shape)
```
