# NumPy ile Veri Analizi FonksiyonlarıBu not defterinde NumPy'nin veri analizi için kullanılan fonksiyonları örneklerle anlatılmaktadır. Fonksiyonlar açıklamalarla birlikte sunulmuş ve yaygın kullanım senaryoları dahil edilmiştir.

```python
import numpy as np
```

## 1. Temel İstatistik Fonksiyonları

```python
data = np.array([10, 20, 30, 40, 50])print("Ortalama:", np.mean(data))      # Aritmetik ortalamaprint("Medyan:", np.median(data))      # Ortanca değerprint("Varyans:", np.var(data))        # Dağılımprint("Standart Sapma:", np.std(data)) # Sapmanın derecesi
```

## 2. Min, Max ve Yüzdelikler

```python
print("Minimum:", np.min(data))print("Maksimum:", np.max(data))print("90. yüzdelik:", np.percentile(data, 90))print("1. Kuantil:", np.quantile(data, 0.25))
```

## 3. Kümülatif İşlemler

```python
print("Toplam:", np.sum(data))print("Kümülatif toplam:", np.cumsum(data))
```

## 4. NaN (Eksik Veri) İşlemleri

```python
data_with_nan = np.array([10, 20, np.nan, 30, 40])print("NaN var mı?:", np.isnan(data_with_nan))print("NaN olmayanlar:", data_with_nan[~np.isnan(data_with_nan)])print("NaN hariç ortalama:", np.nanmean(data_with_nan))
```

## 5. Veriyi Normalize Etme (0-1 arası ölçekleme)

```python
data = np.array([5, 10, 15, 20, 25])normalized = (data - np.min(data)) / (np.max(data) - np.min(data))print("Normalize edilmiş veri:", normalized)
```

## 6. Frekans ve Tekil Değerler

```python
categorical = np.array(['apple', 'banana', 'apple', 'orange', 'banana', 'apple'])unique_vals, counts = np.unique(categorical, return_counts=True)print("Tekil değerler:", unique_vals)print("Tekrar sayıları:", counts)
```
