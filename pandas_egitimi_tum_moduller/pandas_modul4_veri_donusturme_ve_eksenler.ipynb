# 📘 Pandas Eğitimi - MODÜL 4: Veri Dönüştürme ve EksenlerBu modülde `groupby`, `pivot_table`, `apply`, `replace` gibi dönüştürme araçları ve eksen bazlı işlemler anlatılır.

## 🔹 Örnek Veri Seti

```python
import pandas as pdveri = {    "ogrenci": ["Ali", "Ayşe", "Ali", "Mehmet", "Ayşe"],    "ders": ["Matematik", "Fizik", "Fizik", "Matematik", "Kimya"],    "not": [85, 90, 78, 88, 95]}df = pd.DataFrame(veri)df
```

## 🔹 groupby() ile Gruplama

```python
print("Öğrenciye göre not ortalamaları:")print(df.groupby("ogrenci")["not"].mean())
```

## 🔹 apply() ile Fonksiyon Uygulama

```python
def harf_notu(x):    if x >= 90:        return "AA"    elif x >= 80:        return "BB"    elif x >= 70:        return "CC"    else:        return "FF"df["harf"] = df["not"].apply(harf_notu)df
```

## 🔹 replace() ile Değer Değiştirme

```python
df["ogrenci"] = df["ogrenci"].replace({"Ali": "ALİ", "Ayşe": "AYŞE"})df
```

## 🔹 pivot_table ile Özet Tablo Oluşturma

```python
pivot = df.pivot_table(values="not", index="ogrenci", columns="ders", aggfunc="mean")pivot
```

## 🔹 melt() ile Uzun Formata Dönüşüm

```python
df_wide = pivot.reset_index()df_melted = pd.melt(df_wide, id_vars="ogrenci", value_vars=["Fizik", "Kimya", "Matematik"])df_melted
```
