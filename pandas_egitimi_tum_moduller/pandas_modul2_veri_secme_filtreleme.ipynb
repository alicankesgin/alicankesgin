# 📘 Pandas Eğitimi - MODÜL 2: Veri Seçme ve FiltrelemeBu modülde DataFrame yapısından veri seçme, satır/sütun filtreleme ve koşullu seçimler işlenecektir.

## 🔹 Örnek Veri Oluşturma

```python
import pandas as pdveri = {    "isim": ["Ali", "Ayşe", "Mehmet", "Zeynep"],    "yas": [25, 30, 22, 28],    "sehir": ["İzmir", "Ankara", "İstanbul", "Bursa"]}df = pd.DataFrame(veri)df
```

## 🔹 Sütun Seçimi

```python
print("Sadece 'isim' sütunu:")print(df["isim"])print("\nBirden fazla sütun:")print(df[["isim", "yas"]])
```

## 🔹 Satır Seçimi: `iloc` (pozisyona göre)

```python
print("İlk satır:")print(df.iloc[0])print("\nİlk iki satır:")print(df.iloc[:2])
```

## 🔹 Satır ve Sütun Seçimi: `loc` (etikete göre)

```python
df.index = ["a", "b", "c", "d"]print("Satır 'a' ve sütun 'isim':")print(df.loc["a", "isim"])print("\nSatır 'b' ve tüm bilgiler:")print(df.loc["b"])
```

## 🔹 Koşullu Seçim (Boolean Mask)

```python
print("30 yaşından küçük olanlar:")print(df[df["yas"] < 30])print("\nİstanbul'da yaşayanlar:")print(df[df["sehir"] == "İstanbul"])
```
