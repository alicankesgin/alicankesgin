# 📘 Pandas Eğitimi - MODÜL 3: Veri TemizlemeBu modülde eksik verilerle çalışma, tekrarlayanları kaldırma ve veri türü dönüşümleri anlatılır.

## 🔹 Örnek Veri Seti Oluşturma

```python
import pandas as pdimport numpy as npveri = {    "isim": ["Ali", "Ayşe", "Ali", np.nan],    "yas": [25, np.nan, 25, 30],    "sehir": ["İzmir", "Ankara", "İzmir", "Bursa"]}df = pd.DataFrame(veri)df
```

## 🔹 Eksik Veri Kontrolü: `isnull()` ve `notnull()`

```python
print("Hücrelerde eksik veri var mı?")print(df.isnull())print("\nHer sütundaki eksik değer sayısı:")print(df.isnull().sum())
```

## 🔹 Eksik Verileri Silme: `dropna()`

```python
print("Eksik verisi olan satırlar silinir:")print(df.dropna())
```

## 🔹 Eksik Verileri Doldurma: `fillna()`

```python
print("NaN olan 'yas' değerlerini 0 ile doldur:")df_filled = df.copy()df_filled["yas"] = df_filled["yas"].fillna(0)print(df_filled)
```

## 🔹 Yinelenen Satırları Kaldırma: `duplicated()` ve `drop_duplicates()`

```python
print("Yinelenenler:")print(df.duplicated())print("\nTekilleştirilmiş veri:")print(df.drop_duplicates())
```

## 🔹 Veri Tipi Dönüştürme: `astype()`

```python
print("Veri tipleri:")print(df.dtypes)print("\n'Yas' sütununu int'e çevir:")df_tur = df.copy()df_tur["yas"] = df_tur["yas"].fillna(0).astype(int)print(df_tur.dtypes)df_tur
```
