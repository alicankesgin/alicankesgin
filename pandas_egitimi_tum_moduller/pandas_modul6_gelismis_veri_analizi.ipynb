# 📘 Pandas Eğitimi - MODÜL 6: Gelişmiş Veri AnaliziBu modülde veri birleştirme (`concat`, `merge`, `join`), zaman serisi işlemleri ve SQL bağlantısına giriş konuları yer alır.

## 🔹 concat() ile Veri Birleştirme

```python
import pandas as pddf1 = pd.DataFrame({"id": [1, 2], "ad": ["Ali", "Ayşe"]})df2 = pd.DataFrame({"id": [3, 4], "ad": ["Mehmet", "Zeynep"]})birlesik = pd.concat([df1, df2])birlesik
```

## 🔹 merge() ile Anahtara Göre Birleştirme

```python
notlar = pd.DataFrame({"id": [1, 2, 3], "not": [85, 90, 75]})birlesik_merge = pd.merge(birlesik, notlar, on="id", how="left")birlesik_merge
```

## 🔹 join() ile İndeks Bazlı Birleştirme

```python
df_a = pd.DataFrame({"ad": ["Ali", "Ayşe"]}, index=["a", "b"])df_b = pd.DataFrame({"yas": [25, 30]}, index=["a", "b"])df_join = df_a.join(df_b)df_join
```
