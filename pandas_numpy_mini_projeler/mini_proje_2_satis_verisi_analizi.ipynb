## 🔹 Adım 2: Aylık Satış Özeti

## 🔹 Adım 3: Ürün Bazlı Satış Analizi

```python
urun_satis = df.groupby("urun")["toplam"].sum().sort_values(ascending=False)urun_satis.plot(kind="bar", title="Ürün Bazlı Toplam Satış", figsize=(10, 5))
```
