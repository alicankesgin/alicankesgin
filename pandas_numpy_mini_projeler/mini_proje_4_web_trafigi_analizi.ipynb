# 🧪 Mini Proje 4: Web Trafiği AnaliziBu projede bir web sitesine ait günlük ve saatlik trafik verisi simüle edilip analiz edilecektir.

## 🔹 Adım 1: Rastgele Trafik Verisi Üretimi

```python
import pandas as pdimport numpy as npnp.random.seed(42)# Veri oluşturtarih_saatler = pd.date_range(start="2024-01-01", periods=1000, freq="H")veri = {    "zaman": tarih_saatler,    "ziyaretci": np.random.poisson(80, size=1000),  # Ortalama 80 ziyaretçi/saat    "kaynak": np.random.choice(["google", "direct", "email", "referral"], size=1000),    "ulke": np.random.choice(["TR", "US", "DE", "IN"], size=1000)}df = pd.DataFrame(veri)df.head()
```

## 🔹 Adım 2: Günlük Ziyaretçi Analizi

```python
df["tarih"] = df["zaman"].dt.dategunluk = df.groupby("tarih")["ziyaretci"].sum()gunluk.plot(kind="line", title="Günlük Ziyaretçi Sayısı", figsize=(12, 4), grid=True)
```

## 🔹 Adım 3: Saat Bazlı Ortalama Ziyaret

```python
df["saat"] = df["zaman"].dt.hoursaatlik = df.groupby("saat")["ziyaretci"].mean()saatlik.plot(kind="bar", title="Saatlik Ortalama Ziyaretçi", figsize=(10, 4), color="orange")
```

## 🔹 Adım 4: Kaynak Bazlı Ziyaret

```python
kaynaklar = df.groupby("kaynak")["ziyaretci"].sum()kaynaklar.plot(kind="pie", autopct="%1.1f%%", title="Ziyaretçi Kaynak Dağılımı", figsize=(6, 6))
```

## 🔹 Adım 5: Ülke Bazlı Ziyaret

```python
ulke_ziyaret = df.groupby("ulke")["ziyaretci"].sum()ulke_ziyaret.plot(kind="bar", title="Ülke Bazlı Toplam Ziyaretçi", grid=True)
```
