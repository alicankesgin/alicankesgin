# 📘 MODÜL 5: DöngülerDöngüler, belirli işlemleri birden fazla kez otomatik olarak yapmamıza olanak tanır. Bu modülde `for`, `while`, `break`, `continue` ve `pass` ifadeleri incelenecektir.

## 🔹 for DöngüsüBelirli bir koleksiyon üzerinde sırayla gezinmek için kullanılır.

```python
# Liste üzerinde dönmemeyveler = ["elma", "armut", "muz"]for meyve in meyveler:    print("Meyve:", meyve)
```

## 🔹 range() ile for Döngüsürange() belirli bir aralıkta sayı üretir.

```python
for i in range(5):    print("i:", i)# Başlangıç, bitiş ve adım belirlemefor i in range(2, 10, 2):    print("Çift sayı:", i)
```

## 🔹 while DöngüsüKoşul doğru olduğu sürece döngü devam eder.

```python
sayac = 0while sayac < 3:    print("Sayaç:", sayac)    sayac += 1
```

## 🔹 breakDöngüyü tamamen sonlandırır.

```python
for i in range(10):    if i == 5:        break    print("i:", i)
```

## 🔹 continueO anki döngü adımını atlar, bir sonrakine geçer.

```python
for i in range(5):    if i == 2:        continue    print("i:", i)
```

## 🔹 passHenüz tanımlanmamış kod blokları için yer tutucu görevi görür.

```python
for i in range(3):    pass  # Henüz yapılacak bir işlem yok ama sözdizimi gereği boş bırakılmaz
```

## 🔹 Uygulama: 1'den 10'a kadar sayıların karelerini yazdır

```python
for i in range(1, 11):    print(f"{i} sayısının karesi = {i**2}")
```
