## 🔹 Uygulama 1: Not Ortalaması Hesaplama

```python
def ortalama_hesapla():    notlar = []    for i in range(3):        n = float(input(f"{i+1}. notu girin: "))        notlar.append(n)    ort = sum(notlar) / len(notlar)    print("Ortalama:", ort)ortalama_hesapla()
```

## 🔹 Uygulama 2: Basit Hesap Makinesi

```python
def hesap_makinesi():    print("İşlem seçenekleri: +, -, *, /")    a = float(input("1. sayı: "))    b = float(input("2. sayı: "))    islem = input("İşlem türü: ")    if islem == "+":        print("Sonuç:", a + b)    elif islem == "-":        print("Sonuç:", a - b)    elif islem == "*":        print("Sonuç:", a * b)    elif islem == "/":        if b == 0:            print("Sıfıra bölme hatası!")        else:            print("Sonuç:", a / b)    else:        print("Geçersiz işlem!")hesap_makinesi()
```

## 🔹 Uygulama 3: Basit To-Do List

```python
todo_list = []def gorev_ekle(gorev):    todo_list.append(gorev)def listele():    print("Yapılacaklar Listesi:")    for i, g in enumerate(todo_list, 1):        print(f"{i}. {g}")# Örnek kullanımgorev_ekle("Python dersi çalış")gorev_ekle("Alışveriş yap")listele()
```
