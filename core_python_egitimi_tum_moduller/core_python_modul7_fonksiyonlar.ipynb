## 🔹 Parametre Alan Fonksiyon

```python
def topla(a, b):    return a + bsonuc = topla(5, 3)print("Toplam:", sonuc)
```

## 🔹 Varsayılan Parametre Değeri

```python
def selam_ver(isim="Ziyaretçi"):    print("Merhaba", isim)selam_ver()selam_ver("Ali")
```

## 🔹 Anahtar Kelime Argümanları

```python
def ogrenci_bilgi(ad, yas, bolum):    print(f"{ad}, {yas} yaşında, {bolum} bölümünde okuyor.")ogrenci_bilgi(yas=21, ad="Ayşe", bolum="Yazılım")
```

## 🔹 *args (Sınırsız Sayıda Argüman)

```python
def ortalama_hesapla(*notlar):    return sum(notlar) / len(notlar)print("Ortalama:", ortalama_hesapla(70, 80, 90))
```

## 🔹 **kwargs (Sınırsız Anahtar-Değer)

```python
def bilgiler(**kwargs):    for key, value in kwargs.items():        print(f"{key}: {value}")bilgiler(ad="Mehmet", sehir="İzmir", yas=28)
```

## 🔹 Lambda Fonksiyonlar (İsimsiz Fonksiyonlar)

```python
karesi = lambda x: x ** 2print("5'in karesi:", karesi(5))topla = lambda a, b: a + bprint("Toplam:", topla(4, 6))
```

## 🔹 Uygulama: En uzun kelimeyi bulan fonksiyon

```python
def en_uzun_kelime(liste):    return max(liste, key=len)kelimeler = ["elma", "muz", "karpuz", "çilek"]print("En uzun kelime:", en_uzun_kelime(kelimeler))
```
