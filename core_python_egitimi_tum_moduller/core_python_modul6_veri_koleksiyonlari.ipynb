# 📘 MODÜL 6: Veri KoleksiyonlarıPython'da verileri saklamak ve yönetmek için kullanılan temel koleksiyon türleri: list, tuple, set ve dictionary.

## 🔹 Listeler (list)Birden çok veriyi sıralı şekilde saklamak için kullanılır.

```python
meyveler = ["elma", "armut", "muz"]print("Meyveler:", meyveler)# Elemanlara erişimprint("İlk meyve:", meyveler[0])# Eleman eklememeyveler.append("çilek")print("Yeni liste:", meyveler)# Eleman silmemeyveler.remove("armut")print("Armut silindi:", meyveler)# Eleman güncellememeyveler[1] = "kiraz"print("Güncellenmiş liste:", meyveler)
```

## 🔹 Tuple (Demet)Listelere benzer ama değiştirilemez (immutable) yapılardır.

```python
koordinat = (10, 20)print("Koordinat:", koordinat)print("X:", koordinat[0])
```

## 🔹 Set (Küme)Tekil (benzersiz) elemanlar içerir. Sırasızdır.

```python
renkler = {"kırmızı", "mavi", "yeşil"}print("Renkler kümesi:", renkler)# Aynı elemanı tekrar eklemek etkisizdirrenkler.add("mavi")print("Tekrar ekleme:", renkler)# Eleman silmerenkler.discard("yeşil")print("Yeşil silindi:", renkler)
```

## 🔹 Dictionary (Sözlük)Anahtar-değer (key-value) çiftleri saklamak için kullanılır.

```python
ogrenci = {    "ad": "Ali",    "yas": 20,    "bolum": "Bilgisayar"}print("Ad:", ogrenci["ad"])print("Tüm bilgiler:", ogrenci)# Güncellemeogrenci["yas"] = 21print("Yeni yaş:", ogrenci["yas"])# Yeni anahtar eklemeogrenci["mezuniyet"] = 2025print("Güncel sözlük:", ogrenci)
```
