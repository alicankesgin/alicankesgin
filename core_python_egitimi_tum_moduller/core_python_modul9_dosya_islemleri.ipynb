# 📘 MODÜL 9: Dosya İşlemleriBu modülde Python ile dosya açma, okuma, yazma ve güncelleme işlemlerini öğreneceğiz.

## 🔹 Dosya Açma ve Yazma (`write`)

```python
# 'w' modu dosyayı yazar, yoksa oluştururdosya = open("deneme.txt", "w")dosya.write("Merhaba, dosya işlemleri!")dosya.write("Bu ikinci satır.")dosya.close()print("Yazma işlemi tamamlandı.")
```

## 🔹 Dosya Okuma (`read`, `readline`, `readlines`)

```python
# 'r' modu sadece okuma içindirdosya = open("deneme.txt", "r")icerik = dosya.read()print("Dosya içeriği:")print(icerik)dosya.close()
```

```python
dosya = open("deneme.txt", "r")print("Sadece ilk satır:", dosya.readline())dosya.close()
```

```python
dosya = open("deneme.txt", "r")satirlar = dosya.readlines()print("Satırlar listesi:", satirlar)dosya.close()
```

## 🔹 with Bloğu ile Dosya Kullanımı`with` kullanmak dosyanın otomatik olarak kapanmasını sağlar.

```python
with open("deneme.txt", "r") as dosya:    for satir in dosya:        print("Satır:", satir.strip())
```

## 🔹 Dosyaya Ekleme (`append`)

```python
with open("deneme.txt", "a") as dosya:    dosya.write("Bu satır sonradan eklendi.")
```

## 🔹 Dosya Kontrol Uygulaması

```python
try:    with open("deneme.txt", "r") as dosya:        print("Dosya başarıyla okundu.")        print(dosya.read())except FileNotFoundError:    print("Dosya bulunamadı!")
```
