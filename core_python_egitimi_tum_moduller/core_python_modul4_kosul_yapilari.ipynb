# 📘 MODÜL 4: Koşul Yapıları (if-else)Koşul yapıları, program akışını kontrol etmemizi sağlar. Belirli koşullara göre farklı işlemler gerçekleştirebiliriz.

## 🔹 if YapısıKoşul doğruysa (True) kod bloğu çalışır.

```python
x = 10if x > 5:    print("x, 5'ten büyüktür")
```

## 🔹 if-else YapısıKoşul doğruysa bir işlem, yanlışsa başka bir işlem yapılır.

```python
y = 3if y % 2 == 0:    print("y çift sayıdır")else:    print("y tek sayıdır")
```

## 🔹 if-elif-elseBirden fazla durumu kontrol etmek için kullanılır.

```python
not_ortalama = 75if not_ortalama >= 90:    print("Harf notu: AA")elif not_ortalama >= 80:    print("Harf notu: BA")elif not_ortalama >= 70:    print("Harf notu: BB")else:    print("Harf notu: CC veya daha düşük")
```

## 🔹 İç İçe Koşullar (Nested if)

```python
yas = 20uyruk = "TC"if uyruk == "TC":    if yas >= 18:        print("Oy kullanabilirsiniz.")    else:        print("Yaşınız tutmuyor.")else:    print("Yabancı uyruklu seçmensiniz.")
```

## 🔹 Tek Satırlık if-else (Ternary Operatörü)

```python
a = 7sonuc = "Çift" if a % 2 == 0 else "Tek"print("Sayı:", a, "->", sonuc)
```

## 🔹 Uygulama: Sıcaklık Bilgisine Göre Giyim Tavsiyesi

```python
sicaklik = int(input("Hava sıcaklığı kaç derece? "))if sicaklik > 25:    print("Kısa kollu giyebilirsin.")elif sicaklik > 15:    print("Hafif bir mont yeterli.")else:    print("Kalın giyinmeyi unutma.")
```
