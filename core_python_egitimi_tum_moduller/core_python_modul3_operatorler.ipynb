# 📘 MODÜL 3: OperatörlerBu modülde Python'daki temel operatör türlerini öğreneceğiz: aritmetik, karşılaştırma, mantıksal ve atama operatörleri.

## 🔹 Aritmetik OperatörlerPython'da matematiksel işlemler için kullanılır:

```python
a = 10b = 3print("Toplama:", a + b)print("Çıkarma:", a - b)print("Çarpma:", a * b)print("Bölme:", a / b)print("Tam sayı bölme:", a // b)print("Mod (kalan):", a % b)print("Üs alma:", a ** b)
```

## 🔹 Karşılaştırma Operatörleriİki değeri karşılaştırmak için kullanılır. Sonuç `True` veya `False` olur.

```python
x = 5y = 10print("x == y :", x == y)   # Eşit mi?print("x != y :", x != y)   # Eşit değil mi?print("x > y  :", x > y)print("x < y  :", x < y)print("x >= y :", x >= y)print("x <= y :", x <= y)
```

## 🔹 Mantıksal OperatörlerKoşulları birleştirmek için kullanılır: `and`, `or`, `not`

```python
x = 15print("10 < x < 20:", x > 10 and x < 20)print("x < 10 veya x > 20:", x < 10 or x > 20)print("x 10'dan küçük değil:", not x < 10)
```

## 🔹 Atama OperatörleriBir değişkene değer atamak ve üzerinde işlem yapmak için kullanılır.

```python
z = 5z += 3    # z = z + 3print("+= ile arttırma:", z)z *= 2    # z = z * 2print("*= ile çarpma:", z)z -= 4    # z = z - 4print("-= ile azaltma:", z)
```
