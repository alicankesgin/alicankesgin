# 📘 MODÜL 8: Hata YönetimiHatalar programlarda kaçınılmazdır. Python'da hataları kontrol altına almak için `try`, `except`, `finally`, `raise` yapıları kullanılır.

## 🔹 try-except YapısıHata oluşabilecek kodu `try` bloğuna, hata durumunda yapılacak işlemi `except` bloğuna yazarız.

```python
try:    sayi = int(input("Bir sayı girin: "))    print("Girdiğiniz sayı:", sayi)except ValueError:    print("Lütfen geçerli bir sayı girin!")
```

## 🔹 try-except-finally`finally` bloğu her zaman çalışır, hata olsa da olmasa da.

```python
try:    f = open("ornek.txt", "w")    f.write("Dosyaya yazı yazılıyor.")except IOError:    print("Dosya yazılamadı!")finally:    f.close()    print("Dosya kapatıldı.")
```

## 🔹 Birden Fazla Hata Türü Yakalama

```python
try:    x = int(input("Bölünen: "))    y = int(input("Bölen: "))    sonuc = x / y    print("Sonuç:", sonuc)except ValueError:    print("Sayı girilmelidir!")except ZeroDivisionError:    print("Bir sayı 0'a bölünemez!")
```

## 🔹 raise ile Hata FırlatmaKendi hata mesajlarımızı oluşturabiliriz.

```python
def yas_kontrol(yas):    if yas < 0:        raise ValueError("Yaş negatif olamaz!")    else:        print("Yaş geçerli.")yas_kontrol(25)# yas_kontrol(-5)  # Bu satır çalıştırılırsa hata fırlatır
```

## 🔹 Uygulama: Kullanıcıdan yaş al ve kontrol et

```python
try:    yas = int(input("Yaşınızı girin: "))    if yas < 0:        raise ValueError("Yaş negatif olamaz.")    print("Yaşınız:", yas)except ValueError as hata:    print("Hata oluştu:", hata)
```
