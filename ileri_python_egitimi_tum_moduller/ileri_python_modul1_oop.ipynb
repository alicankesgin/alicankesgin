# 🧠 İleri Python - MODÜL 1: Nesne Tabanlı Programlama (OOP)Bu modülde Python'da sınıflar, nesneler, metodlar, kalıtım ve özel metodlar gibi OOP kavramları ele alınacaktır.

## 🔹 Sınıf Tanımı ve `__init__()` Metodu

```python
class Araba:    def __init__(self, marka, model):        self.marka = marka        self.model = model    def bilgi(self):        print(f"Marka: {self.marka}, Model: {self.model}")arac1 = Araba("Toyota", "Corolla")arac1.bilgi()
```

## 🔹 Nesne Özellikleri (Attributes) ve Metodlar

```python
class Ogrenci:    def __init__(self, ad, notu):        self.ad = ad        self.notu = notu    def goster(self):        print(f"{self.ad} adlı öğrencinin notu: {self.notu}")ali = Ogrenci("Ali", 85)ali.goster()
```

## 🔹 Kalıtım (Inheritance)

```python
class Canli:    def __init__(self, isim):        self.isim = isim    def ses_cikar(self):        print(f"{self.isim} bir ses çıkardı.")class Kedi(Canli):    def ses_cikar(self):        print(f"{self.isim} miyavladı.")minnoş = Kedi("Minnoş")minnoş.ses_cikar()
```

## 🔹 Çok Biçimlilik (Polymorphism)

```python
def hayvan_sesi(hayvan):    hayvan.ses_cikar()class Kopek:    def ses_cikar(self):        print("Hav hav")class Kus:    def ses_cikar(self):        print("Cik cik")hayvan_sesi(Kopek())hayvan_sesi(Kus())
```

## 🔹 Özel Metodlar (`__str__`, `__len__`, `__eq__`)

```python
class Kitap:    def __init__(self, ad, sayfa):        self.ad = ad        self.sayfa = sayfa    def __str__(self):        return f"{self.ad} kitabı, {self.sayfa} sayfa"    def __len__(self):        return self.sayfa    def __eq__(self, other):        return self.sayfa == other.sayfak1 = Kitap("Python101", 300)k2 = Kitap("AI Giriş", 300)print(k1)print("Sayfa sayısı:", len(k1))print("Kitaplar eşit mi?:", k1 == k2)
```
