# 📘 İleri Python - MODÜL 4: datetime ile Zaman ve Tarih İşlemleriBu modülde Python'da tarih, saat, zaman farkı ve biçimlendirme işlemleri anlatılır.

## 🔹 datetime ile Şu Anki Zamanı Alma

```python
from datetime import datetimesimdi = datetime.now()print("Şu an:", simdi)print("Yıl:", simdi.year)print("Ay:", simdi.month)print("Gün:", simdi.day)print("Saat:", simdi.hour)print("Dakika:", simdi.minute)
```

## 🔹 Zamanı Formatlama (`strftime`)

```python
tarih = datetime.now()print("Tarih (Yıl-Ay-Gün):", tarih.strftime("%Y-%m-%d"))print("Saat (24 saat formatı):", tarih.strftime("%H:%M"))print("Tarih-Saat (tam):", tarih.strftime("%d/%m/%Y %H:%M:%S"))
```

## 🔹 Yazıdan Tarih Oluşturma (`strptime`)

```python
tarih_str = "2024-05-01 14:30"tarih_obj = datetime.strptime(tarih_str, "%Y-%m-%d %H:%M")print("Yazıdan oluşturulan tarih:", tarih_obj)
```

## 🔹 Zaman Farkı Hesaplama (`timedelta`)

```python
from datetime import timedeltabugun = datetime.now()gelecek = bugun + timedelta(days=7)print("Bugün:", bugun)print("7 gün sonra:", gelecek)fark = gelecek - bugunprint("Zaman farkı (gün):", fark.days)
```

## 🔹 Sadece Tarih veya Saat Nesneleri

```python
from datetime import date, timetarih = date.today()saat = time(14, 45)print("Sadece tarih:", tarih)print("Sadece saat:", saat)
```
