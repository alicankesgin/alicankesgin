## 🔹 Karakter Sınıfları ve Özel Semboller

```python
metin = "Email adresim: ornek123@mail.com ve digeri test.mail@demo.org"# E-posta adresi yakalamaemail_deseni = r"[\w.-]+@[\w.-]+"emailler = re.findall(email_deseni, metin)print("Bulunan emailler:", emailler)
```

## 🔹 Örnek: Telefon Numarası Bulma

```python
metin = "İletişim: 0532-123-4567 veya 0212-555-1234"tel_deseni = r"\d{3,4}-\d{3}-\d{4}"telefonlar = re.findall(tel_deseni, metin)print("Telefonlar:", telefonlar)
```
