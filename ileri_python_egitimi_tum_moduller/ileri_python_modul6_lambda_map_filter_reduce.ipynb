# 📘 İleri Python - MODÜL 6: Lambda, Map, Filter, ReduceBu modülde Python'da fonksiyonel programlama araçları olan `lambda`, `map`, `filter`, `reduce` anlatılacaktır.

## 🔹 Lambda FonksiyonlarLambda fonksiyonlar, kısa süreli ve tek satırlık işlevler tanımlamak için kullanılır.

```python
karesi = lambda x: x ** 2print("5'in karesi:", karesi(5))
```

## 🔹 map() FonksiyonuBir fonksiyonu listedeki her elemana uygular.

```python
sayilar = [1, 2, 3, 4, 5]kareler = list(map(lambda x: x ** 2, sayilar))print("Kareler:", kareler)
```

## 🔹 filter() FonksiyonuBir fonksiyonla koşul sağlayan elemanları süzer.

```python
sayilar = [10, 15, 20, 25, 30]ciftler = list(filter(lambda x: x % 2 == 0, sayilar))print("Çift sayılar:", ciftler)
```

## 🔹 reduce() FonksiyonuBir listedeki tüm elemanları tek bir değere indirger.

```python
from functools import reducesayilar = [1, 2, 3, 4]toplam = reduce(lambda x, y: x + y, sayilar)print("Toplam:", toplam)
```

## 🔹 map vs filter vs reduce| Fonksiyon | Amaç | Sonuç ||-----------|------|--------|| `map()`   | Dönüştürme | Tüm elemanlara işlem uygular || `filter()`| Filtreleme | Şartı sağlayanları seçer || `reduce()`| İndirgeme | Listeyi tek değere indirger |
