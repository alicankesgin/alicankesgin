## 🔹 Modül Nedir?Modül, `.py` uzantılı ve içinde Python kodları bulunan dosyalardır. Bir modüldeki fonksiyonları veya sınıfları başka bir dosyada kullanmak mümkündür.

## 🔹 Dahili (Built-in) Modül Kullanımı

````python
import mathprint("Pi sayısı:", math.pi)print("Karekök:", math.sqrt(25))
````

## 🔹 from-import ile Fonksiyon veya Sınıf Alma

````python
from math import pow, factorialprint("2 üzeri 3:", pow(2, 3))print("5 faktöriyel:", factorial(5))
````

## 🔹 `__name__ == '__main__'` KullanımıBir modül hem doğrudan çalıştırılabilir hem de başka bir yerden çağrıldığında çalışması kontrol edilebilir.

````python
# demo.py dosyasında şu kod olsun:def selam():    print("Merhaba!")if __name__ == "__main__":    selam()```Bu dosya doğrudan çalıştırılırsa `selam()` fonksiyonu çağrılır ama başka bir dosyadan import edilirse çağrılmaz.
````

## 🔹 Paket Nedir?Birden fazla modülü bir klasörde toplamak ve `__init__.py` dosyasıyla modülleri yönetmek mümkündür. Bu yapıya Python'da 'paket' denir.```my_package/├── __init__.py├── mod1.py└── mod2.py```Kullanım:```pythonfrom my_package import mod1```
