# 📘 İleri Python - MODÜL 3: JSON, CSV ve XML Veri İşlemeBu modülde Python ile JSON, CSV ve XML dosyaları üzerinden veri okuma/yazma işlemleri örneklerle anlatılır.

## 🔹 JSON İşlemleriPython sözlükleri ile uyumlu olan JSON, veri serileştirme ve paylaşımı için sıkça kullanılır.

```python
import json# Python sözlüğüveri = {    "ad": "Ali",    "yas": 25,    "sehir": "İzmir"}# Sözlüğü JSON formatına çevirjson_str = json.dumps(veri, indent=4)print("JSON biçimi:", json_str)# Dosyaya yazwith open("veri.json", "w") as dosya:    json.dump(veri, dosya)# Dosyadan okuwith open("veri.json", "r") as dosya:    okunan = json.load(dosya)    print("Dosyadan okunan veri:", okunan)
```

## 🔹 CSV Dosyaları ile ÇalışmakCSV (Virgülle Ayrılmış Değerler), tabular veri için yaygın biçimdir.

```python
import csv# Yazmawith open("ornek.csv", mode="w", newline="") as dosya:    yazici = csv.writer(dosya)    yazici.writerow(["Ad", "Soyad", "Yaş"])    yazici.writerow(["Ali", "Yılmaz", 30])    yazici.writerow(["Ayşe", "Kaya", 25])# Okumawith open("ornek.csv", mode="r") as dosya:    okuyucu = csv.reader(dosya)    for satir in okuyucu:        print("Satır:", satir)
```
