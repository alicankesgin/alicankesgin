## 🔹 Closure (Kapanış Fonksiyonları)Bir fonksiyonun, başka bir fonksiyonu döndürmesi ve dış kapsamdaki değişkeni hatırlaması.

```python
def selam_ver(mesaj):    def yaz():        print("Mesaj:", mesaj)    return yazmerhaba = selam_ver("Merhaba!")merhaba()
```

## 🔹 Generator Fonksiyonlar (`yield`)Bellek dostu veri üretimi için kullanılır.

```python
def sayac():    for i in range(3):        yield ifor sayi in sayac():    print("Üretilen:", sayi)
```

```python
def kareler(n):    for i in range(n):        yield i ** 2for k in kareler(5):    print("Karesi:", k)
```

## 🔹 Decorator FonksiyonlarBir fonksiyonun davranışını değiştirmek için kullanılır.

```python
def logla(fonksiyon):    def sarici():        print(f"{fonksiyon.__name__} çağrılıyor...")        fonksiyon()        print("Çağrı bitti.")    return sarici@logladef merhaba():    print("Merhaba Dünya!")merhaba()
```

## 🔹 Parametreli Fonksiyonlarda Decorator Kullanımı

```python
def logla(fonk):    def sarici(*args, **kwargs):        print(f"{fonk.__name__} fonksiyonu çağrılıyor...")        sonuc = fonk(*args, **kwargs)        print("Çağrı tamamlandı.")        return sonuc    return sarici@logladef topla(a, b):    return a + bprint("Sonuç:", topla(3, 4))
```
